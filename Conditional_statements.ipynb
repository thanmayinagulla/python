```python
11.#Create a simple calculator (+, -, *, /).
a_str = input("Enter the first value:")
b_str = input("Enter the second value:")
operation = input("Enter the operation (+, -, *, /):")

try:
    a = float(a_str)
    b = float(b_str)
except ValueError:
    print("Invalid input. Please enter numbers for values.")
else:
    if operation == '+':
        print(f"Result: {a + b}")
    elif operation == '-':
        print(f"Result: {a - b}")
    elif operation == '*':
        print(f"Result: {a * b}")
    elif operation == '/':
        if b != 0:
            print(f"Result: {a / b}")
        else:
            print("Error: Cannot divide by zero.")
    else:
        print("Invalid operation. Please enter +, -, *, or /.")
```
